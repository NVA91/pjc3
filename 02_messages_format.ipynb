# 02 — Messages Format

Dieses Notebook erklärt das `messages`-Format der Claude API und zeigt, wie Konversationen aufgebaut werden.

## Basic Setup

```python
from dotenv import load_dotenv
from anthropic import Anthropic

# Umgebungsvariable laden
load_dotenv()

# Sucht automatisch nach ANTHROPIC_API_KEY in der .env
client = Anthropic()
```

## Einfache Nachricht senden

```python
response = client.messages.create(
    model="claude-haiku-4-5-20251001",
    max_tokens=1000,
    messages=[
        {"role": "user", "content": "What flavors are used in Dr. Pepper?"}
    ]
)

print(response)
```

## Warum das Messages-Format?

Das `messages`-Format ermöglicht **Kontexterhaltung**: Claude hat Zugriff auf die gesamte Konversationshistorie — sowohl User- als auch Assistant-Nachrichten — und kann dadurch kohärentere und relevantere Antworten generieren.

> **Hinweis:** Viele Anwendungsfälle benötigen gar keine Konversationshistorie. Es ist völlig in Ordnung, eine Liste mit nur einer einzigen Nachricht zu übergeben!

## Das `messages`-Format

Der `messages`-Parameter erwartet eine Liste von Dictionaries. Jedes Dictionary hat:

| Key | Wert |
|-----|------|
| `role` | `"user"` oder `"assistant"` |
| `content` | String oder Liste von Content-Dictionaries |

> **Wichtig:** `user` und `assistant` müssen sich immer abwechseln.

```python
# Einzelne Nachricht
messages = [
    {"role": "user", "content": "Hello Claude! How are you today?"}
]
```

```python
# Mehrere Nachrichten — vollständige Konversation
messages = [
    {"role": "user",      "content": "Hello Claude! How are you today?"},
    {"role": "assistant", "content": "Hello! I'm doing well, thank you. How can I assist you today?"},
    {"role": "user",      "content": "Can you tell me a fun fact about ferrets?"},
    {"role": "assistant", "content": "Sure! Did you know that excited ferrets make a clucking vocalization known as 'dooking'?"},
]

print(f"{len(messages)} Nachrichten, abwechselnd user/assistant:") 
for m in messages:
    print(f"  [{m['role']}] {m['content'][:60]}...")
```
